```python
import torch
from torch.autograd import Variable as V
import torch.nn.functional as F
```

```python
x_data = V(torch.Tensor([[1.0], [2.0], [3.0], [4.0]]))
y_data = V(torch.Tensor([[0.], [0.], [1.0], [1.]]))
           
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1, 1) # 1 in (from x value), 1 out
    
    def forward(self, x):
        y_hat = F.sigmoid(self.linear(x))
        return y_hat
```

```python
model = Model()
criterion = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(1000):
    y_hat = model(x_data) # forward pass
    
    loss = criterion(y_hat, y_data)
    print(epoch, loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step() # update
```

```python
tester_var = V(torch.Tensor([[1.0]]))
print("predict (after training)", model(tester_var).data[0][0] > 0.5)

tester_var = V(torch.Tensor([[7.0]]))
print("predict (after training)", model(tester_var).data[0][0] > 0.5)
```
