```python
def mse(y_hat, y):
    return ((y_hat-y)**2).mean()
```

```python
t = torch.tensor([5.,5,5])
a = torch.nn.Parameter(t); a
```

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
```

```python
lr = 1e-1
for t in range(100): update()
```
